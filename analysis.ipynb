```python
import matplotlib.pyplot as plt
from image_analyzer import triple_gaussian

# Prepare data for all three analyzers
# (assumes plot_survival_vs_parameter() was already called for each)
analyzers = {
    'Shallow Trap': shallow_analyzer,
    'Medium Trap': medium_analyzer,
    'Deep Trap': deep_analyzer,
}

colors = {'Deep Trap': 'blue', 'Shallow Trap': 'green', 'Medium Trap': 'orange'}
fit_colors = {'Deep Trap': 'darkblue', 'Shallow Trap': 'darkgreen', 'Medium Trap': 'darkorange'}

# Create 3 subplots side by side
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

for idx, (name, analyzer) in enumerate(analyzers.items()):
    ax = axes[idx]
    
    x_data = np.array(analyzer.parameters)
    y_data = np.array(analyzer.survival_ratios)
    
    # Filter out None values
    valid_mask = np.array([y is not None for y in y_data])
    x_fit = x_data[valid_mask]
    y_fit = np.array([y for y in y_data if y is not None])
    
    # Plot data points
    ax.scatter(x_fit / 1e6, y_fit, color=colors[name], label='Data', alpha=0.7, s=30)
    
    # Use stored fit parameters (no need to redo the fit!)
    if hasattr(analyzer, 'fit_params') and analyzer.fit_params is not None:
        popt = analyzer.fit_params
        
        # Plot fit curve using stored parameters
        x_smooth = np.linspace(x_fit.min(), x_fit.max(), 500)
        y_smooth = triple_gaussian(x_smooth, *popt)
        ax.plot(x_smooth / 1e6, y_smooth, color=fit_colors[name], linewidth=2, label='Fit')
    else:
        print(f"No fit parameters found for {name}. Run plot_survival_vs_parameter() first.")
    
    ax.set_xlabel('MW Frequency (MHz)', fontsize=10)
    ax.set_ylabel('Survival Ratio', fontsize=10)
    ax.set_title(name, fontsize=12)
    ax.legend(loc='best')
    ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Print splitting (difference between outer Gaussian centers)
print("\n" + "="*60)
print("Splitting (Center of Gaussian 3 - Center of Gaussian 1)")
print("="*60)

for name, analyzer in analyzers.items():
    if not hasattr(analyzer, 'fit_params') or analyzer.fit_params is None:
        print(f"\n{name}: No fit data available")
        continue
        
    popt = analyzer.fit_params
    perr = analyzer.fit_errors
    
    # Sort to get extremities (min and max centers)
    # Centers are at indices 1, 4, 7
    centers = sorted([(popt[1], perr[1]), (popt[4], perr[4]), (popt[7], perr[7])])
    c_min, c_min_err = centers[0]
    c_max, c_max_err = centers[2]
    
    splitting = c_max - c_min
    splitting_err = np.sqrt(c_min_err**2 + c_max_err**2)
    
    print(f"\n{name}:")
    print(f"  Left center:  ({c_min/1e6:.4f} ± {c_min_err/1e6:.4f}) MHz")
    print(f"  Right center: ({c_max/1e6:.4f} ± {c_max_err/1e6:.4f}) MHz")
    print(f"  Splitting:    ({splitting/1e3:.2f} ± {splitting_err/1e3:.2f}) kHz")
```
